# Creación del Cerebro de la I.A.

Se pretende utlizar tres capas ocultas, las cuales cada una tiene 18, 36 y 27 neuronas respectivamente, utilizando la funcion de activacion sigmoid.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
#Cerebro
def build_model():
  model = keras.Sequential([
    layers.Dense(18, activation='sigmoid', input_shape=[3]),   #input_shape  se ingresa cantida de Xi
    layers.Dense(36, activation='sigmoid'),
    layers.Dense(27, activation='sigmoid'),
    layers.Dense(1)
  ])
  optimizer = tf.keras.optimizers.RMSprop(learning_rate=0.0001)
  model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse'])
  return model
```

```python
#Se crea el modelo Cerebro (modelo de redes neuronales)
model = build_model()
```

# Entrenamiento de la Red Neuronal

Se pretende utlizar un EPOCHS de 10000

```python
import tensorflow_docs as tfdocs
import tensorflow_docs.modeling
```

```python
EPOCHS = 10000
```

```python
history = model.fit(
    X,Y,
    epochs=EPOCHS, verbose=0,
    callbacks=[tfdocs.modeling.EpochDots()])
```
